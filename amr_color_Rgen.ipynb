```python
# Step 1: Initialize coarse grid, domain [0,1]
# We have mx cells, indexed by 0 to mx-1, and x[i] is the mid-point of cell i
mxc = 100
dx = 1./mxc
xc = linspace(dx/2, 1.-dx/2, mxc)  # computational grid

tfinal = 1
dt = dx
nsteps = int(tfinal/dx)
#dt = float(tfinal)/nsteps

meqn = 1  # number of equations in system

# intital condition
q0 = where(abs(xc) < 0.5,  1., 0.)

# Coefficient function u
def u_fnc(x):
    #u = where(abs(x) < 0.5,  exp(x), 1.)
    u = where(abs(x) < 0.5,  x**2, 0.1)
    #n = len(x)
    #n = numpy.array([x]).size
    #u = zeros(n)
    #for i in range(n):
    #    u[i]= 0.75
    
    return u
```

```python
# Step 2: Solve problem for ONE time step on coarse grid, for time step dt and 2*dt
# u is the coefficient function in the form of a vector where u[i] = u(x) evaluated at x[i]
# Similarly, q0 is a vector with the initial condition evaluated at cell mid-points x[i]
# THE SOLUTION INCLUDES TWO GHOST CELLS ON EACH SIDE

def color_onestep(u,q0,dt,dx):
    
    mx = len(q0)
    qn = hstack([0,0, q0, 0,0])   # add two ghost cells on each end
                                  # qn[2] through qn[mx+1]
    cfl = abs(u)*dt/dx
    
    # Extrapolate boundary conditions (linear extrapolation)
    qn[1] = 2*qn[2] - qn[3]
    qn[0] = 2*qn[1] - qn[2]
    qn[mx+2] = 2*qn[mx+1] - qn[mx]
    qn[mx+3] = 2*qn[mx+2] - qn[mx+1]
    
    qnp = zeros(qn.shape)   # initialize array
    flux = zeros(qn.shape)
    
    for i in range(2,mx+2):
        flux[i] = 0.5*abs(u[i-2])*(1-(dt/dx)*abs(u[i-2]))*(qn[i]-qn[i-1])
    
    flux[1] = 2*flux[2] - flux[3]
    flux[0] = 2*flux[1] - flux[2]
    flux[mx+2] = 2*flux[mx+1] - flux[mx]
    flux[mx+3] = 2*flux[mx+2] - flux[mx+1]
    
    #waves = zeros(meqn,mx+4)
    #for i in range(1,mx+4):
    #    waves[:,i] = qn[:,i] - qn[:,i-1]
        

        
    for i in range(2,mx+2):
        if u[i-2] > 0:
            qnp[i] = qn[i] - (dt/dx) * u[i-2]*(qn[i]-qn[i-1])
        else:
            qnp[i] = qn[i] - (dt/dx) * u[i-2]*(qn[i+1]-qn[i])
            
        qnp[i] = qnp[i]-(dt/dx)*(flux[i+1]-flux[i])
        
    #qnew = qnp[2:(mx+2)]   # throw away ghost cells
    # Extrapolate boundary conditions
    qnp[1] = 2*qnp[2] - qnp[3]
    qnp[0] = 2*qnp[1] - qnp[2]
    qnp[mx+2] = 2*qnp[mx+1] - qnp[mx]
    qnp[mx+3] = 2*qnp[mx+2] - qnp[mx+1]
    
    return qnp
```

```python
def bilin_interp(x_l,x_r,y_b,y_t,ql_b,qr_b,ql_t,qr_t,in_x,in_y):
    value = (1/((x_r-x_l)*(y_t-y_b)))*(ql_b*(x_r-in_x)*(y_t-in_y) + qr_b*(in_x-x_l)*(y_t-in_y) + ql_t*(x_r-in_x)*(in_y-y_b) + qr_t*(in_x-x_l)*(in_y-y_b))
    return value
```

```python
def amr_color(u_fnc,q0,xc,tfinal,nsteps):
    # q0 = initial condition evaluated on coarse grid 
    # u_fnc = a function name evaluating the coefficient function u at the grid points
    # xc = coarse grid, domain [a,b] is divided into mx_c cells, and xc[i] is the mid-point of cell i
        
    dx = xc[1]- xc[0]             # length of cells in (given) coasrse grid
    dt = float(tfinal)/nsteps     # length of time-step
    
    u = u_fnc(xc)
    # u = velocity (coefficient) function, u[i] is cell-centered value in cell [i],
    #     currently evaluated at cell-midpoint
    
    mx_c = len(xc)              # number of grid cells in coarse grid
    
    qn = hstack([0,0,q0,0,0])
    qn[1] = 2*qn[2] - qn[3]
    qn[0] = 2*qn[1] - qn[2]
    qn[mx_c+2] = 2*qn[mx_c+1] - qn[mx_c]
    qn[mx_c+3] = 2*qn[mx_c+2] - qn[mx_c+1]
    
    # Step 1: initalize coarse grid values q0 and u: Done before calling this function
    # The rest of the steps are done over nsteps/K time steps, where K is the the period for regridding
    for istep in range(nsteps):
        
        # Step 2: Solve problem on coarse grid for single time step
        #         Done by calling a function 'color_onestep', which advances the solution on a grid
        #         with spacing dx, over a single time step of length dt, returns solution with two ghost cell
        #         values on each side
        
        qnp = color_onestep(u,qn[2:mx_c+2],dt,dx)
    
        # Step 3: Estimate error and flag cells based on Richardson extrapolation
        #         We only allow for single level refinement.
        # This estimation and grid generation procedure is only performed every K time steps. Buffer zone size and K
        # are inter-related.
        
        p = 3
        R = 2**p # refinement factor, chosen as a power of 2
        dx_hat = dx/R 
        dt_hat = dt/R
        
        # Compare errors between solving problem on coarse grid for 2 time steps of size 2*dt_hat 
        # and one step of size 4*dt_hat
        qnew_D = color_onestep(u,qn[2:mx_c+2],2*dt_hat,dx)[2:mx_c+2] # returned solutions had ghost cells
        qnew_D = color_onestep(u,qnew_D[2:mx_c+2],2*dt_hat,dx)[2:mx_c+2]
        qnew_2D = color_onestep(u,qn[2:mx_c+2],4*dt_hat,dx)[2:mx_c+2]
        
        # Error
        error = zeros(mx_c)
        s = 2             # order of accuracy
        flag_cnt = 0      # flagged cell count
        tol = (dx**(s+2))/100

        for i in range(0,mx_c):
            error[i] = (qnew_D[i]-qnew_2D[i])/(2**(s+1)-2)     
            if abs(error[i]) >= tol:
                flag_cnt = flag_cnt + 1
        
        # Use AMR only if any cells are flagged for refinement!
        if flag_cnt > 0:
            
            # Cell flagging
            flagged = numpy.empty(flag_cnt)  # contains indices of cells flagged for refinement
            
            flag_cnt = 0
            for i in range(0,mx_c):
                if abs(error[i]) >= tol:
                    flagged[flag_cnt] = i
                    flag_cnt = flag_cnt + 1
            print flagged
   
            # cluster generation
            bfs = 2 # buffer-size
            gc = 1  # number of ghost cells
            cls_tol = 2*(bfs+gc) # threshold for clustering
    
            temp = numpy.empty(2*flag_cnt) # will be storing start and end points of clusters here
            idx = 0
            temp[idx] = max(flagged[0]-bfs,0.) # lower end of first cluster
            clstr_cnt = 1

            for i in range(0,flag_cnt-1):
                if (flagged[i+1]-flagged[i]) > cls_tol:
                    temp[idx+1] = flagged[i] + bfs  # upper end of (s/2 + 1)th cluster
                    clstr_cnt = clstr_cnt + 1
                    idx = idx+2
                    temp[idx] = flagged[i+1] - bfs

            temp[idx+1] = min(flagged[flag_cnt-1]+ bfs, mx_c-1)
            clstr_cnt = int(clstr_cnt)

            clusters = [0 for i in xrange(2*clstr_cnt)]

            for i in range(0,2*clstr_cnt):
                clusters[i] = int(temp[i])
            #print clusters
    
            # Generate fine grid
            
            # Look for the largest sized cluster, to create array qhat of appropriate size
            cellnum = [0 for i in xrange(clstr_cnt)]  # Array to store number of cells in each cluster, sans ghost cells.
            
            for i in range(0,clstr_cnt):
                cellnum[i] = int(clusters[(2*i)+1] - clusters[2*i]) +1
            maxclsize = max(cellnum)
   
            qhat = numpy.empty([clstr_cnt, R*maxclsize+4])  # include two ghost cells (on fine grid) on either side of each cluster
            
            xhat = numpy.empty([clstr_cnt, R*maxclsize+4])
        
            corr_l = zeros(clstr_cnt)   # store correction terms for fix up step
            corr_r = zeros(clstr_cnt)
    
        
            for iclstr in range(0,clstr_cnt):
                
                clidx = range(clusters[(2*iclstr)]-1, clusters[(2*iclstr)+1]+2)
                # vector of cellnum + 2 coarse cell indices in cluster iclstr
                # (could start at -1, if the first cell in the first cluster is 0, which is fine
                # since we already have ghost cell values stored in qn and qnp; in fact, qn[1] and qnp[1]
                # correspond to function values in cell '-1' on coarse grid. The value for cell j is stored in 
                # qn[j+2] and qnp[j+2])
                
                mxf_clstr = R*(cellnum[iclstr]) + 4   # number of fine grid cells in cluster 
                
                # cell midpoints for fine grid
                # Left fine grid ghost-cells (may be out of domain)
                xhat[iclstr,0] = xc[clidx[1]] - (dx/(2*R))*(3+R)
                xhat[iclstr,1] = xc[clidx[1]] - (dx/(2*R))*(1+R)
                
                # Right ghost cells
                xhat[iclstr,mxf_clstr-2] = xc[clidx[cellnum[iclstr]]] + (dx/(2*R))*(R+1)
                xhat[iclstr,mxf_clstr-1] = xc[clidx[cellnum[iclstr]]] + (dx/(2*R))*(R+3)
                
                # Interior fine grid cells
                for j in range(0,cellnum[iclstr]):
                    for icell in range(0,R):
                        xhat[iclstr,R*j + 2 + icell] = xc[clidx[j+1]] + (dx/(2*R))*(2*icell + 1 - R)
                        
                
                # Initializing fine grid q values (including ghost cells)
                # Left ghost cells
                slope1 = (qn[clidx[0]+2] - qn[clidx[0]+1])/dx
                slope2 = (qn[clidx[0]+3] - qn[clidx[0]+2])/dx
                if (slope1*slope2)<= 0:
                    slope = 0
                elif slope1 > 0:
                    slope = min(slope1,slope2)
                else:
                    slope = max(slope1,slope2)
                qhat[iclstr, 0] = qn[clidx[0]+2] + slope*(xc[clidx[1]]-dx - xhat[iclstr, 0])
                qhat[iclstr, 1] = qn[clidx[0]+2] + slope*(xc[clidx[1]]-dx - xhat[iclstr, 1])
                
                
                # Right ghost cells
                slope1 = (qn[clidx[cellnum[iclstr]]+3] - qn[clidx[cellnum[iclstr]]+2])/dx
                slope2 = (qn[clidx[cellnum[iclstr]]+4] - qn[clidx[cellnum[iclstr]]+3])/dx
                if (slope1*slope2)<= 0:
                    slope = 0
                elif slope1 > 0:
                    slope = min(slope1,slope2)
                else:
                    slope = max(slope1,slope2)
                qhat[iclstr, mxf_clstr-2] = qn[clidx[cellnum[iclstr]]+3] + slope*(xc[clidx[cellnum[iclstr]]]+dx - xhat[iclstr, mxf_clstr-2])
                qhat[iclstr, mxf_clstr-1] = qn[clidx[cellnum[iclstr]]+3] + slope*(xc[clidx[cellnum[iclstr]]]+dx - xhat[iclstr, mxf_clstr-1])
                
                # Interior fine grid cells
                for j in range(0,cellnum[iclstr]):
                    slope1 = (qn[clidx[j+1]+2] - qn[clidx[j+1]+1])/dx
                    slope2 = (qn[clidx[j+1]+3] - qn[clidx[j+1]+2])/dx
                    if (slope1*slope2)<= 0:
                        slope = 0
                    elif slope1 > 0:
                        slope = min(slope1,slope2)
                    else:
                        slope = max(slope1,slope2)
                    # minmod slope. can use other slopes, limiters, etc
                    
                    for icell in range(R):
                        qhat[iclstr,R*j +2 +icell] = qn[clidx[j+1]+2] + slope*(xc[clidx[j+1]] - xhat[iclstr,R*j +2 +icell] )
                    
                uhat = u_fnc(xhat[iclstr,0:mxf_clstr])
            
                for tstep in range(0,R-1):
                                    
                    # compute correction for this fine grid time step
                    # Fix-up in ghost cells to ensure conservation
                    # Left ghost cell, indexed by clidx[0]. Look at the fine grid cell closest to the coarse-fine grid
                    # interface, given by xhat[iclstr,1]
                    
                    u_lgc = uhat[1]
                    # if u_lgc >= 0: # Right going correction, nothing to do
                    if u_lgc < 0: # Left going correction
                        corr_l[iclstr] = corr_l[iclstr] + (dt_hat/dx) * u_lgc*(qhat[iclstr,1] - qnp[clidx[0]+2]) 
            
                    # Right ghost cell, indexed by clidx[cellnum[iclstr]+1]. Look at the fine grid cell closest to the 
                    # coarse-fine grid interface, given by xhat[iclstr,mxf_clstr-2]
                    u_rgc = uhat[mxf_clstr-2]
            
                    if u_rgc >= 0: # Right going correction
                        corr_r[iclstr] = corr_r[iclstr] + (dt_hat/dx) * u_rgc*(qnp[clidx[cellnum[iclstr]]+3]- qhat[iclstr,mxf_clstr-2])
                    
                    # Compute qhat for time t_n + dt_hat
                    qhat[iclstr,0:mxf_clstr] = color_onestep(uhat,qhat[iclstr,0:mxf_clstr],dt_hat,dx_hat)[2:mxf_clstr+2]
            
                    
                    # bilin_interp(x_l,x_r,y_b,y_t,ql_b,qr_b,ql_t,qr_t,in_x,in_y)
                    # The way this is done depends on R. If R<=2 i.e. p < 2, need to look at two coarse grid ghost cells, 
                    # otherwise one suffices
                    
                    if R <= 2:
                        qhat[iclstr,0] = bilin_interp(xc[clidx[1]]-2*dx, xc[clidx[1]]-dx, 0, dt, qn[clidx[0]+1],
                                                      qn[clidx[0]+2], qnp[clidx[0]+1], qnp[clidx[0]+2],
                                                      xhat[iclstr,0], (tstep+1)*dt_hat)
                                       
                        qhat[iclstr,mxf_clstr-1] = bilin_interp(xc[clidx[cellnum[iclstr]]]+dx,
                                                                xc[clidx[cellnum[iclstr]]]+2*dx, 0, dt,
                                                                qn[clidx[cellnum[iclstr]]+3], qn[clidx[cellnum[iclstr]]+4],
                                                                qnp[clidx[cellnum[iclstr]]+3], qnp[clidx[cellnum[iclstr]]+4] ,
                                                                xhat[iclstr,mxf_clstr-1], (tstep+1)*dt_hat)
                    else:
                        qhat[iclstr,0] = bilin_interp(xc[clidx[1]]-dx, xc[clidx[1]], 0, dt, qn[clidx[0]+2],
                                                      qn[clidx[1]+2], qnp[clidx[0]+2], qnp[clidx[1]+2],
                                                      xhat[iclstr,0], (tstep+1)*dt_hat)
    
                        qhat[iclstr,mxf_clstr-1] = bilin_interp(xc[clidx[cellnum[iclstr]]], xc[clidx[cellnum[iclstr]]]+dx,
                                                                0, dt, qn[clidx[cellnum[iclstr]]+2],
                                                                qn[clidx[cellnum[iclstr]]+3], qnp[clidx[cellnum[iclstr]]+2],
                                                                qnp[clidx[cellnum[iclstr]]+3],
                                                                xhat[iclstr,mxf_clstr-1], (tstep+1)*dt_hat)
    
    
                    qhat[iclstr,1] = bilin_interp(xc[clidx[1]]-dx, xc[clidx[1]], 0, dt, qn[clidx[0]+2],
                                                      qn[clidx[1]+2], qnp[clidx[0]+2], qnp[clidx[1]+2],
                                                      xhat[iclstr,1], (tstep+1)*dt_hat)
    
                    qhat[iclstr,mxf_clstr-2] = bilin_interp(xc[clidx[cellnum[iclstr]]], xc[clidx[cellnum[iclstr]]]+dx,
                                                                0, dt, qn[clidx[cellnum[iclstr]]+2],
                                                                qn[clidx[cellnum[iclstr]]+3], qnp[clidx[cellnum[iclstr]]+2],
                                                                qnp[clidx[cellnum[iclstr]]+3],
                                                                xhat[iclstr,mxf_clstr-2], (tstep+1)*dt_hat)
                    
                # Solve over another time step dt_hat to get fine grid values for time t_n+1
                qhat[iclstr,0:mxf_clstr] = color_onestep(uhat,qhat[iclstr,0:mxf_clstr],dt_hat,dx_hat)[2:mxf_clstr+2]   
                    
                # Add corrections to function values in the ghost cells at time t_(n+1)
                qnp[clidx[0]+2] = qnp[clidx[0]+2] + corr_l[iclstr]
                qnp[clidx[cellnum[iclstr]]+3] = qnp[clidx[cellnum[iclstr]]+3] + corr_r[iclstr]
            
                # Average to get coarse grid values within fine grid region
                for k in range(0,cellnum[iclstr]):
                    a = 0
                    for icell in range(0,R):
                        a = a + qhat[iclstr, R*k +icell + 2]
                        
                    qnp[clidx[k+1]+2] = a/R
        
        
        # updated values for next time step iteration
        qn = qnp
        
        plt.figure(istep) 
        plot(xc,qn[2:mx_c+2])
        ylim([-0.5,1.5])
        plt.show()
    
    return qn[2:mx_c+2]
```

```python
def no_amr_color(u_fnc,q0,xc,tfinal,nsteps):
    # q0 = initial condition evaluated on coarse grid (cell averages)
    # u_fnc = a function name evaluating the coefficient function u at the grid points
    # u = velocity (coefficient) function, u[i] is cell-centered value in cell [i],
    #     currently evaluated at cell-midpoint
    # xc = coarse grid, domain [a,b] is divided into mx cells, and xc[i] is the mid-point of cell i
        
    dx = xc[1]- xc[0]             # length of cells in (given) coasrse grid
    dt = float(tfinal)/nsteps   # length of time-step
    u = u_fnc(xc)
    
    mx_c = len(xc)              # number of grid cells in coarse grid
    
    # Step 1: initalize coarse grid values q0 and u: Done before calling this function
    qn = q0
    
    for istep in range(nsteps):
        
    
        # Step 2: Solve problem on coarse grid for single time step
        #         Done by calling a function 'color_onestep', which advances the solution on a given grid
        #         with spacing dx, over a single time step of length dt, returns solution with two ghost cell
        #         values on each side
        qnp = color_onestep(u,qn,dt,dx)
        qn = qnp[2:mx_c+2]
        
        plt.figure(istep) 
        plot(xc,qn)
        ylim([-0.5,1.5])
        plt.show()
        
    return qn
```
